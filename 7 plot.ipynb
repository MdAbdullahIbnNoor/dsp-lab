```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sequence x(n) with starting index -5
x = np.array([1, 2, 3, 4, 5, 6, 7, 6, 5, 4, 3, 2, 1])
n = np.arange(-5, len(x) - 5)  # n starts from -5 to len(x) - 6

# Define the sequence y(n) = 2x(n - 5) - 3x(n + 4)
y = np.zeros_like(x)

for i in range(len(x)):
    if i - 5 >= 0 and i + 4 < len(x):
        y[i] = 2 * x[i - 5] - 3 * x[i + 4]

# Plotting
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.stem(n,x)
plt.title(r'Sequence $x(n)$')
plt.xlabel(r'$n$')
plt.ylabel(r'$x(n)$')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.stem(n, y)
plt.title(r'Sequence $y(n) = 2x(n-5) - 3x(n+4)$')
plt.xlabel(r'$n$')
plt.ylabel(r'$y(n)$')
plt.grid(True)

plt.tight_layout()
plt.show()
```
