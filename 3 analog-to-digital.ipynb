```python
import numpy as np 
import matplotlib.pyplot as plt

t=np.arange(0,2*np.pi,0.001)
fa=int(input())
aa=2
ysignal=aa*np.sin(2*np.pi*fa*t)
plt.plot(t,ysignal)
plt.grid()
sampling_rate=int(input())
```

```python
def analog_signal(f,a,x):
    y=a*np.sin(2*np.pi*f*x)
    return y
```

# Sampling

```python
x=np.arange(0,2,1/sampling_rate)
y=analog_signal(fa,aa,x)
plt.title("SAMPLING")
plt.stem(x,y)
```

# Quantize

```python
def quantize(x):
    y=np.zeros(len(x))
    #for i in range(len(x)):
      #  y[i]=int(x[i])
    y=np.round(x).astype(int)
    return y
```

```python
y=quantize(y)
print(y)
```

```python
plt.title("QUANTIZED VALUES")
plt.xlabel("Continuous Amplitude")
plt.ylabel("Discrete Time")
plt.stem(x,y)
```

# Coding

```python
def coding(x):
    code=""
    for i in range(len(x)):
        temp=bin(x[i])
        if(x[i]<0):
            temp=temp[3:]
            while(len(temp)<3):
                temp='0'+temp
            temp='1'+temp
        else:
            temp=temp[2:]
            while(len(temp)<4):
                temp='0'+temp
        code=code+temp
    return code
```

```python
code=coding(y)
```

```python
x=np.arange(len(code))
y=np.zeros(len(code))
for i in range(len(code)):
    if(code[i]=='1'):
        y[i]=1
    else:
        y[i]=0
```

```python
plt.step(x,y,where="post")
plt.yticks([-aa,aa])
plt.title("DIGITAL SIGNAL")
plt.xlabel("Amplitude")
plt.ylabel("Time")
plt.grid()
plt.show()
```
