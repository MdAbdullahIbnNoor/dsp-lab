```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
sampling_rate = 100  # Sampling rate in Hz
T = 1.0 / sampling_rate  # Sampling interval
N = 1000  # Number of sample points
k = np.arange(N)  # Time vector

# Define the signal
f_k = (0.25 +
       2 * np.sin(2 * np.pi * 5 * k * T) +
       np.sin(2 * np.pi * 12.5 * k * T) +
       1.5 * np.sin(2 * np.pi * 20 * k * T) +
       0.5 * np.sin(2 * np.pi * 35 * k * T))

# Compute the FFT
f_fft = np.fft.fft(f_k)
f_freqs = np.fft.fftfreq(N, T)

# Only positive frequencies
positive_freqs = f_freqs[:N // 2]
positive_magnitude = np.abs(f_fft)[:N // 2] * 2 / N

# Plot the signal in time domain
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(k * T, f_k)
plt.title('Signal in Time Domain')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')

# Plot the frequency spectrum
plt.subplot(2, 1, 2)
plt.stem(positive_freqs, positive_magnitude, basefmt=" ")
plt.title('Frequency Spectrum')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude')
plt.tight_layout()

# Display the plots
plt.show()
```
