```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import firwin, freqz

# Define filter specifications
fp1 = 0.2  # Passband edge frequency 1
fp2 = 0.35  # Passband edge frequency 2
fs1 = 0.1  # Stopband edge frequency 1
fs2 = 0.425  # Stopband edge frequency 2
M = 32  # Filter length
window = 'hamming'  # Hamming window

# Calculate normalized frequencies
fp1_norm = fp1
fp2_norm = fp2
fs1_norm = fs1
fs2_norm = fs2

# Design the FIR filter using firwin function
h = firwin(M, [fs1_norm, fp1_norm, fp2_norm, fs2_norm], pass_zero=False, window=window)

# Compute the frequency response of the filter
w, freq_response = freqz(h, worN=8000)

# Plot the frequency response
plt.figure(figsize=(10, 6))
plt.plot(0.5 * w / np.pi, 20 * np.log10(np.abs(freq_response)), 'b')
plt.axvline(fp1, color='g', linestyle='--', label=f'Passband Edge 1: {fp1} Hz')
plt.axvline(fp2, color='r', linestyle='--', label=f'Passband Edge 2: {fp2} Hz')
plt.axvline(fs1, color='orange', linestyle='--', label=f'Stopband Edge 1: {fs1} Hz')
plt.axvline(fs2, color='purple', linestyle='--', label=f'Stopband Edge 2: {fs2} Hz')
plt.title('Bandpass FIR Filter Frequency Response')
plt.xlabel('Normalized Frequency')
plt.ylabel('Gain (dB)')
plt.ylim(-80, 5)
plt.grid(True)
plt.legend()

plt.show()
```
