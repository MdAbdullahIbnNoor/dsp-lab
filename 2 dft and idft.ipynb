```python
import numpy as np
import matplotlib.pyplot as plt

# Function to compute the DFT
def DFT(x):
    N = len(x)
    X = np.zeros(N, dtype=complex)
    for k in range(N):
        for n in range(N):
            X[k] += x[n] * np.exp(-2j * np.pi * k * n / N)
            
    return X

# Function to compute the IDFT
def IDFT(X):
    N = len(X)
    x = np.zeros(N, dtype=complex)
    for n in range(N):
        for k in range(N):
            x[n] += X[k] * np.exp(2j * np.pi * k * n / N)
            
    return x / N

# User input for the length of the sequence
N = int(input("Enter the length of the sequence N: "))

# User input for the sequence values
print(f"Enter {N} values for the sequence:")
x = np.array([int(input(f"x[{i}]: ")) for i in range(N)])


# Compute the DFT of the signal
X = DFT(x)

# Compute the IDFT of the transformed signal
x_reconstructed = IDFT(X)

# Plot the original signal
plt.figure(figsize=(16, 10))
plt.subplot(3, 1, 1)
plt.stem(np.arange(N), x, 'b', markerfmt="bo", basefmt="-b")
plt.title('Original Signal')
plt.xlabel('Sample index')
plt.ylabel('Amplitude')

# Plot the magnitude spectrum of the DFT
plt.subplot(3, 1, 2)
plt.stem(np.arange(N), np.abs(X), 'r', markerfmt="ro", basefmt="-r")
plt.title('Magnitude Spectrum of DFT')
plt.xlabel('Frequency bin')
plt.ylabel('Magnitude')

# Plot the reconstructed signal using IDFT
plt.subplot(3, 1, 3)
plt.stem(np.arange(N), x_reconstructed.real, 'g', markerfmt="go", basefmt="-g")
plt.title('Reconstructed Signal from IDFT')
plt.xlabel('Sample index')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```
